* Which languages are the most people interested in learning over the next year?

```python
#print(data['LanguageWantToWorkWith'].head())
d = data['LanguageWantToWorkWith'].str.split(';', expand = True)
d = d.stack()
#print(d.head(10))
d = d.value_counts()
print(d.head())
d = d.reset_index()
d.columns = ['Languages', 'counts']

language = list(d['Languages'])
count = list(d['counts'])

#d.plot.bar()                                 # bar chart by pandas
fig = plt.figure(figsize = (15, 5))
plt.bar(language, count, color ='maroon',width = 0.4)

plt.xticks(rotation = 75)
plt.xlabel("languages")
plt.ylabel("Interested")
plt.title("People interested in Languages")
plt.show()
```

 * Which Databases are the most people interested in learning over the next year?

```python
#print(data['DatabaseWantToWorkWith'].head())

def databaseFunction(data):
    
    res = data.to_frame()
    col = []
    # Iterate over the column
    for idx, value  in data.iteritems():
        for option in value.split(';'):
            if not option in res.columns:
                col.append(option)
                res[option] = False
            # Mark the value in the option column as True
            res.at[idx, option] = True
    return res[col]

dB = data.dropna(subset=['DatabaseWantToWorkWith'])
result = databaseFunction(dB.DatabaseWantToWorkWith)

print(result.head())
```

* How much does remote work matter to employment? (2022)

```python
#print(data['RemoteWork'].unique())
work = data.dropna(subset = ['RemoteWork'])
work = work['RemoteWork'].value_counts().reset_index()
work.columns = ["RemoteWork", "Count"]
print(work['Count'])

fig = plt.figure(figsize = (15, 5))
plt.pie(list(work['Count']), labels = list(work['RemoteWork']), autopct = '%1.1f%%')

#plt.title("People working model (2022)")
plt.show()
```

* Are you more likely to get a job as developer if you've a master's degree?

```python
#print(data[['Employment', 'EdLevel']].head())
#print(data['Employment'].unique())
#print(data['EdLevel'].unique())

data = data.dropna(subset = ['Employment', 'EdLevel'])
totalEmployed = len(data[data['Employment'].str.contains('Employed')])
data = (data[['Employment', 'EdLevel']] [data.Employment.str.contains('Employed', case = True) & (data.EdLevel.str.contains('Masters degree', case = False))])
totalEmployedMaster = len(data)
total = [totalEmployed, totalEmployedMaster]
lab = ['total employed', 'employed with master']

plt.xlabel("Employment Label")
plt.ylabel("People")
plt.bar(lab, total, width = 0.4)
plt.show()
```

* Gender of coders

```python
gender = data.dropna(subset = ['Gender'])
#print(data.Gender.unique())
males = gender['Gender'].eq("Man").sum()
females = gender['Gender'].eq("Woman").sum()
others = len(gender['Gender']) - (males + women)

print("Males: ", males , ' Females: ', females , ' Others: ' , others)

plt.pie([males, females, others], labels = ["Males", "Females", "Others"], autopct = '%1.1f%%')
plt.show()
```
